```python
import requests
import pandas as pd
from datetime import datetime, timedelta

api_key = "95be546de5af4fe7b67211140232503"
```

```python
def fetch_weather_data(start_date, end_date, location, api_key):
    base_url = f"https://api.worldweatheronline.com/premium/v1/past-weather.ashx"

    params = {
        "q": f"{location},India",
        "format": "json",
        "date": start_date,
        "enddate": end_date,
        "key": api_key,
        "tp": 3
    }

    response = requests.get(base_url, params=params)
    print(f"Response status code: {response.status_code}")
    print(f"Response content: {response.content}")

    data = response.json()

    weather_data = []

    for day_data in data["data"]["weather"]:
        date = day_data["date"]
        for hourly_data in day_data["hourly"]:
            weather_data.append({
                "date": date,
                "time": hourly_data["time"],
                "tempC": hourly_data["tempC"],
                "humidity": hourly_data["humidity"],
                "windspeedKmph": hourly_data["windspeedKmph"],
            })

    return pd.DataFrame(weather_data)
```

```python
start_date = "2015-01-01"
end_date = "2022-12-31"
location = "Delhi"

weather_df = fetch_weather_data(start_date, end_date, location, api_key)
print(weather_df.head())
```
